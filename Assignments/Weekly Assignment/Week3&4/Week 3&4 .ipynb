```python
# Load libraries
import pandas as pd
import xlrd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import numpy as np
```

```python
# Read "world-population" data file
world_population_df = pd.read_excel('world-population.xlsm')
world_population_df.head()
```

# Charts 
## - Line Chart
- Use seaborn
- Use matplotlib
- Use Pandas

```python
# Use seaborn to plot line chart 
import seaborn as sns
sns.lineplot(data=world_population_df, x="Year", y="Population")
```

```python
# Use matplotlib to plot line charts 
import matplotlib.pyplot as plt
plt.plot(world_population_df["Year"], world_population_df["Population"])
```

```python
# Use Pandas to plot line charts
import pandas as pd
world_population_df.plot.line(x="Year", y="Population")
```

## - Step Chart

```python
# Use matplotlib ti plot Step Chart 
import numpy as np

plt.step(world_population_df["Year"], world_population_df["Population"])
plt.show()
```

```python
# Use matplotlib ti plot Step Chart
import numpy as np

plt.step(world_population_df["Year"], world_population_df["Population"], 'g^', where='pre')
# ‘g^’ denotes the line color is green and the marker should be an upward triangle.
plt.show()
```
